```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# Initialize the webdriver (using Chrome in this case)
driver = webdriver.Chrome()

# Define the base URL for the draft
#base_url = "https://www.footballdb.com/draft/draft.html?lg=NFL&yr=1994&rnd="
base_url = "https://www.footballdb.com/draft/draft.html?lg=NFL&"

# Initialize an empty list to store the data
draft_data = []

for year in range(1994, 1999):
    
    # Loop through rounds 1 to 7
    for rnd in range(1, 8):
        # Construct the URL for each round
        url = f"{base_url}yr={year}&rnd={rnd}"
        print(url)

        # Load the page for the current round
        driver.get(url)

        # Wait for the page content to load
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, 'tr')))

        # Give it a moment to fully load (you can adjust this time if necessary)
        time.sleep(2)

        # Get page source
        page_source = driver.page_source

        # Use BeautifulSoup to parse the page
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find all rows in the draft table (rows with class 'tr')
        rows = soup.find_all('div', class_='tr')

        # Extract data from each row
        for row in rows:
            try:
                # Round and Pick (both are in divs with the same class)
                test = row.find_all('div', class_='td w10 m15')
                # skip empty row
                if len(test) == 0:
                    continue
                #print(test)
                #round_col = row.find('div', class_='td w10 m15')
                #pick_col = row.find('div', class_='td w10 m15')
                round_col = test[0]
                pick_col = test[1]

                #test2 = row.find_all('a')
                #print(test2)
                # Team (now a <b> tag)
                team_col = row.find('b')

                # Player (now an <a> tag)
                player_col = row.find('div', class_='td w20 td-clear hidden-xs')
                #player_col= test2[1]
                #college_col = test2[2]
                # Position (unchanged)
                position_col = row.find('div', class_='td w10 td-clear hidden-xs')

                # College (now a <b> tag, similar to Team)
                college_col = row.find('div', class_='td w25 hidden-xs')

                # Check if all required columns are found
                if round_col and pick_col and team_col and player_col and position_col and college_col:
                    draft_data.append([
                        year,
                        round_col.get_text(strip=True),
                        pick_col.get_text(strip=True),
                        team_col.get_text(strip=True),
                        player_col.get_text(strip=True),
                        position_col.get_text(strip=True),
                        college_col.get_text(strip=True)
                    ])
                else:
                    print(f"Skipping incomplete row in round {rnd}.")
            except Exception as e:
                print(f"Error processing row in round {rnd}: {e}")

# Check if data is being extracted
if draft_data:
    # Create DataFrame
    draft_df = pd.DataFrame(draft_data, columns=["Year", "Round", "Pick", "Team", "Player", "Position", "College"])

    # Debugging: Print the first few rows
    print(draft_df.head())

    # Save to CSV
    draft_df.to_csv('draft_data_2023.csv', index=False)
else:
    print("No data found.")

# Close the driver
driver.quit()
```
