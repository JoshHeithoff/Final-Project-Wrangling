```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Import the combined dataframe (assuming the combined data is saved here)
combined_df = pd.read_csv('combined_fp_data.csv')

# Step 2: Filter data for Wide Receiver (WR) position
wr_df = combined_df[combined_df['Position'] == 'WR']

# Step 3: Create a box plot for 'FantPt' across all rounds for WR
plt.figure(figsize=(12, 6))
sns.boxplot(x='Round', y='FantPt', data=wr_df, palette='Set2')

# Step 4: Add labels for the average, min, and max for each round
for i in range(len(wr_df['Round'].unique())):
    round_data = wr_df[wr_df['Round'] == wr_df['Round'].unique()[i]]['FantPt']
    avg = round_data.mean()
    min_val = round_data.min()
    max_val = round_data.max()

    # Adding text labels to the box plot
    plt.text(i, avg, f'Avg: {avg:.2f}', horizontalalignment='center', verticalalignment='center', fontsize=10, color='black')
    plt.text(i, min_val, f'Min: {min_val:.2f}', horizontalalignment='center', verticalalignment='center', fontsize=10, color='black')
    plt.text(i, max_val, f'Max: {max_val:.2f}', horizontalalignment='center', verticalalignment='center', fontsize=10, color='black')

# Step 5: Add labels and title
plt.title('Fantasy Points for WR Position by Round', fontsize=16)
plt.xlabel('Round', fontsize=14)
plt.ylabel('Fantasy Points (FantPt)', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Import the combined dataframe (assuming the combined data is saved here)
combined_df = pd.read_csv('combined_fp_data.csv')

# Step 2: Clean the data (convert 'FantPt' to numeric if necessary)
combined_df['FantPt'] = pd.to_numeric(combined_df['FantPt'], errors='coerce')

# Step 3: Filter the data to include only QB, RB, WR, TE positions
filtered_df = combined_df[combined_df['Position'].isin(['QB', 'RB', 'WR', 'TE'])]

# Step 4: Group by 'Position' and calculate the average 'FantPt' across all rounds
avg_fantpt_by_position = filtered_df.groupby('Position')['FantPt'].mean().reset_index()

# Step 5: Sort the data by average fantasy points (optional for better visualization)
avg_fantpt_by_position = avg_fantpt_by_position.sort_values('FantPt', ascending=False)

# Step 6: Create a bar plot to visualize the average fantasy points by position
plt.figure(figsize=(10, 6))
sns.barplot(x='Position', y='FantPt', data=avg_fantpt_by_position, palette='viridis')

# Step 7: Add labels for each bar (average FantPt value for each position)
for index, row in avg_fantpt_by_position.iterrows():
    plt.text(index, row['FantPt'] + 0.2, f"{row['FantPt']:.2f}", ha='center', fontsize=12)

# Step 8: Add plot details
plt.title('Average Fantasy Points by Position (QB, RB, WR, TE) Across All Rounds', fontsize=16)
plt.xlabel('Position', fontsize=14)
plt.ylabel('Average Fantasy Points (FantPt)', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Step 9: Display the plot
plt.tight_layout()
plt.show()
```
