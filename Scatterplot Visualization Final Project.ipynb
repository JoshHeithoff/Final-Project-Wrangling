```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Import the combined dataframe and any other required dataframes
combined_df = pd.read_csv('combined_fp_data.csv')  # Assuming this is where your combined dataframe is stored

# Step 2: Create a scatterplot for 'FantPt' vs 'PPR'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='FantPt', y='PPR', data=combined_df, color='purple', alpha=0.7)

# Adding plot details
plt.title('Scatterplot of Fantasy Points (FantPt) vs PPR Points', fontsize=16)
plt.xlabel('Fantasy Points (FantPt)', fontsize=14)
plt.ylabel('PPR Points', fontsize=14)

# Step 3: Display the plot
plt.tight_layout()
plt.show()
```
