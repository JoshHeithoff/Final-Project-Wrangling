```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the combined dataframe (assuming the combined data is saved here)
combined_df = pd.read_csv('combined_fp_data.csv')

# Step 2: Group by 'Round' and calculate the average 'FantPt' for each round
avg_fantpt_per_round = combined_df.groupby('Round')['FantPt'].mean().reset_index()

# Step 3: Create a line chart for the average 'FantPt' per round
plt.figure(figsize=(12, 6))
plt.plot(avg_fantpt_per_round['Round'], avg_fantpt_per_round['FantPt'], marker='o', color='b')

# Step 4: Add labels for the average score at each round
for i in range(len(avg_fantpt_per_round)):
    plt.text(avg_fantpt_per_round['Round'].iloc[i], 
             avg_fantpt_per_round['FantPt'].iloc[i], 
             f"{avg_fantpt_per_round['FantPt'].iloc[i]:.2f}", 
             ha='center', va='bottom', fontsize=10)

# Step 5: Add labels and title
plt.title('Average Fantasy Points (FantPt) by Round', fontsize=16)
plt.xlabel('Round', fontsize=14)
plt.ylabel('Average FantPt', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```
