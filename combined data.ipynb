```python
import pandas as pd  # Importing pandas for working with dataframes and CSVs

# Load the scraped fantasy stats CSV (this file should already exist from previous scraping)
df = pd.read_csv('nfl_fantasy_stats_2000.csv')

# Load the draft data CSV (from 2023 data)
draft_df = pd.read_csv('draft_data_2023.csv')
```

```python
# Merge the two DataFrames on the 'Player' column
combined_df = pd.merge(df, draft_df, on='Player', how='outer')  # 'outer' keeps all rows from both DataFrames

# Save the combined data to a new CSV file titled 'combined_fp_data.csv'
combined_df.to_csv('combined_fp_data.csv', index=False)

# Debugging: Print the first few rows of the combined DataFrame
print(combined_df.head())
```
